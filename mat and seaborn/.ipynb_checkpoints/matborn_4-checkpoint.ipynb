Create a 2D NumPy array of shape (10, 3) filled with random integers between 1 and 100.
Think of it like 10 students with 3 subject marks. ðŸŽ“ðŸ“š

Then:

Using Matplotlib, plot a grouped bar chart (each student on X-axis, 3 bars for their marks).

Using Seaborn, plot the same as a barplot (hint: you might need to reshape the NumPy array before plotting).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
array = np.random.randint(1,100,30).reshape(10,3)
```

```python
array
```

```python
x = np.arange(array.shape[0])
```

```python
x
```

```python
width = 0.25
```

```python
plt.bar(x-width,array[:,0],width,label="Sub1")
plt.bar(x,array[:,1],width,label="Sub2")
plt.bar(x+width,array[:,2],width,label="Sub3")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.xticks(x, labels=[f"Student {i+1}" for i in x],rotation=90)
plt.title("Student Marks by Subject")
plt.legend()
```
