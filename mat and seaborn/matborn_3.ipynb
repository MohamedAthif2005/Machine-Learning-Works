Generate a 1D NumPy array of 100 random numbers from a normal distribution with:

mean = 50

standard deviation = 10

Then:

Plot a histogram using Matplotlib (plt.hist).

Plot the same data using Seabornâ€™s distplot / histplot with KDE curve.

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
array = np.random.normal(loc=50,scale=10,size=100)
array
```

```python
plt.hist(array,bins=10)
```

```python
sns.histplot(array,kde=True)
```
